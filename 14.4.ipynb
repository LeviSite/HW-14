```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("BRCA_pam50.tsv", sep="\t", index_col=0)

df2 = df.iloc[:, :50].loc[:, ["ESR1", "PGR", "ERBB2", "MKI67"]].melt(var_name="Gene", value_name="Expression")

sns.boxplot(x="Gene", y="Expression", hue = list(df["Subtype"])*4, data=df2)
plt.tight_layout()
plt.show()
```
